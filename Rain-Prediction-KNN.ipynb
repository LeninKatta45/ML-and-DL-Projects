```python
import pandas as pd
df=pd.read_csv("testset.csv")
```

```python
df.head()
```

```python
x=df[[" _conds"," _dewptm"," _fog"," _hum"," _thunder"]]
```

```python
x.head()
```

```python
y=df[" _rain"]
```

```python
s=pd.get_dummies(x[" _conds"])
```

```python
s
```

```python
j=list(s.columns)
```

```python
j
```

```python
X=pd.concat([x,s],axis=1)
```

```python
X.drop([" _conds"],axis=1,inplace=True)
```

```python
X
```

```python
y
```

```python
X[[" _dewptm"]]=X[[" _dewptm"]].fillna(X[" _dewptm"].mean())
```

```python
X[[" _hum"]]=X[[" _hum"]].fillna(X[" _hum"].mean())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
s=KNeighborsClassifier()
s.fit(x_train,y_train)
```

```python
s.score(x_test,y_test)
```
